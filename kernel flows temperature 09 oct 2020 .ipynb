```python
from scipy.spatial.distance import pdist, squareform

def poly_fit(x, y, degree, fit="RANSAC"):
  # check if we can use RANSAC
  if fit == "RANSAC":
    try:
      # ignore ImportWarnings in sklearn
      with warnings.catch_warnings():
        warnings.simplefilter("ignore", ImportWarning)
        import sklearn.linear_model as sklin
        import sklearn.preprocessing as skpre
    except ImportError:
      warnings.warn(
        "fitting mode 'RANSAC' requires the package sklearn, using"
        + " 'poly' instead",
        RuntimeWarning)
      fit = "poly"

  if fit == "poly":
    return np.polyfit(x, y, degree)
  elif fit == "RANSAC":
    model = sklin.RANSACRegressor(sklin.LinearRegression(fit_intercept=False))
    xdat = np.asarray(x)
    if len(xdat.shape) == 1:
      # interpret 1d-array as list of len(x) samples instead of
      # one sample of length len(x)
      xdat = xdat.reshape(-1, 1)
    polydat = skpre.PolynomialFeatures(degree).fit_transform(xdat)
    try:
      model.fit(polydat, y)
      coef = model.estimator_.coef_[::-1]
    except ValueError:
      warnings.warn(
        "RANSAC did not reach consensus, "
        + "using numpy's polyfit",
        RuntimeWarning)
      coef = np.polyfit(x, y, degree)
    return coef
  else:
    raise ValueError("invalid fitting mode ({})".format(fit))
```

```python
y=np.stack((zzz1, zzz2, zzz3), axis=-1)    
n,d = y.shape # n: lenght of the data; d = 3 # dimension of the data

# if data is not in right form as length x dimension, transpose it
if n < d:
    y = y.T
    n,d = y.shape 

# some parameters, please check nolds documentation
# you might need to work on them to find the best way to choose it 
min_tsep=10; trajectory_len=13

# distance matrix between points in phase space
dists = squareform(pdist(y))
for i in range(n):
    dists[i, max(0, i - min_tsep):i + min_tsep + 1] = float("inf")
    
ntraj = n - trajectory_len + 1
nb_idx = np.argmin(dists[:ntraj, :ntraj], axis=1)

```

```python
# build divergence trajectory by averaging distances along the trajectory
# over all neighbor pairs
div_traj = np.zeros(trajectory_len, dtype=float)
for k in range(trajectory_len):
    # calculate mean trajectory distance at step k
    indices = (np.arange(ntraj) + k, nb_idx + k)
    div_traj_k = dists[indices]
    # filter entries where distance is zero (would lead to -inf after log)
    nonzero = np.where(div_traj_k != 0)

    if len(nonzero[0]) == 0:
        # if all entries where zero, we have to use -inf
        div_traj[k] = -np.inf
    else:
        div_traj[k] = np.mean(np.log(div_traj_k[nonzero]))
# filter -inf entries from mean trajectory
ks = np.arange(trajectory_len)
finite = np.where(np.isfinite(div_traj))
ks = ks[finite]
div_traj = div_traj[finite]
```

```python
def lyap_exp(f, dim, args=None, t_start = 0, t_end=1000, t_step=0.1, dt=0.001, x_0 = [0.0, 1.0, 0.0], Transient=0, silent=0):
    '''
    Compute Lyapunov Exponent of $f$
    '''
    
    t = t_start
    timesteps = int(round((t_end-t_start)/t_step))
    if silent is 0 : print('Calculating Lyapunov Exponent | steps = %d'%timesteps)
    y, y1 = np.zeros(dim *(dim+1)), np.zeros(dim *(dim+1)) 
    cum, gsc, znorm = np.zeros(dim), np.zeros(dim), np.zeros(dim)
    y[:dim] = x_0
    y[dim:] = np.eye(dim).flatten() 
    #Control = np.zeros(shape=(timesteps,dim+1))
    
    idx = np.arange(dim,dim*(dim+1),dim) # index numbers for the vectorized diagonals of the Jacobian

    
    for lya_iter in range(timesteps):
        if silent is 0 and lya_iter%10 == 0 : print('Progress: %.2f '%(100 * lya_iter /timesteps), end='\r')
        sol = solve_ivp(f, [t,t+t_step], y0=y, method='RK45',args=args)
        Y = sol.y # solutions
        
        t += t_step
        y = Y[:,-1]
        #Control[lya_iter,:] = [t,Y[0,-1],Y[1,-1],Y[2,-1]]
        
        # transpose
        y1[dim:]= y[dim:].reshape(dim,dim).T.flatten()
        
        ## gram-schmidt         
        znorm[0] = LA.norm(y1[idx]) 
        y1[idx]/=znorm[0]

        for j in range(1,dim):
            for k in range(j): # compute Gram-Schmidt Coeffs
                gsc[k]= np.sum(y1[idx + j] * y1[idx + k])
            for l in range(j): # New vector
                y1[idx + j] -= gsc[l] * y1[idx + l]        
                    
            znorm[j] = LA.norm(y1[idx+j]) # compute the norm for off-diagonal elements
            y1[idx+j]/=znorm[j]
       
        # transpose back
        y[dim:]= y1[dim:].reshape(dim,dim).T.flatten()
        
        if (t>Transient):
            cum += np.log(znorm)
           
    return cum/(t-t_start-Transient)#, Control
```

```python
def lorenz(t,x,sig,rho,beta):
    return [sig*(x[1]-x[0]), -x[0]*x[2]+rho*x[0]-x[1], x[0]*x[1]-beta*x[2]]
```

```python
def lin_lorenz(t,x,sig,rho,beta):
    Y = x[3:].reshape(3,3).T
    f = np.zeros(12)
    
    f[:3] = lorenz(t,x,sig,rho,beta)

    # Jacobian matrix
    J=np.array([[    -sig,  sig,     0],
                [rho-x[2],   -1,    -x[0]], 
                [    x[1], x[0], -beta]])

    f[3:] = J.dot(Y).T.reshape(-1)
    return f
```
