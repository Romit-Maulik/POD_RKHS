```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcdefaults()
```

```python
mpl.rcParams['lines.linewidth'] = 3
mpl.rcParams['font.sans-serif'] = 'Arial'
mpl.rcParams['text.color'] = '#909090'
mpl.rcParams['axes.labelcolor']= '#909090'
mpl.rcParams['xtick.color'] = '#909090'
mpl.rcParams['ytick.color'] = '#909090'
mpl.rcParams['font.size'] = 18
mpl.rcParams['legend.fontsize'] = 18
mpl.rcParams['axes.labelsize'] = 24
```

```python
# Load the true test fields
test_true = np.load('sst_var_test',allow_pickle=True)[1318:1486]
# Load the mask
mask = np.load('mask',allow_pickle=True)

for week in range(168):
    test_true[week] = test_true[week]*mask[0]
    
# Load the HYCOM fields
test_hycom = np.nan_to_num(np.load('HYCOM_Data_Interpolated.npy'),nan=-5)

# Load the Predicted fields
test_pred = np.load('RKHS_Predicted_Fields.npy')[1318:1486]

# Load CESM fields
test_cesm = np.load('CESM_Test_Fields.npy')[1318:1486]
for week in range(168):
    test_cesm[week] = test_cesm[week]*mask[0]
```

```python
num_weeks = np.arange(np.shape(test_true)[0])
```

```python
plt.figure(figsize=(12,6))
plt.plot(num_weeks[:],test_pred[:,85,210],label='Predicted')
plt.plot(num_weeks[:],test_hycom[:,85,210],label='HYCOM')
plt.plot(num_weeks[:],test_cesm[:,85,210],label='CESM')
plt.plot(num_weeks[:],test_true[:,85,210],label='True')
plt.ylabel('Celsius')
plt.xlabel('Week')
plt.ylim((22,32))
plt.legend(ncol=4)
plt.tight_layout()
plt.savefig('NOAA_Probe_85_210.png')
plt.show()
```
